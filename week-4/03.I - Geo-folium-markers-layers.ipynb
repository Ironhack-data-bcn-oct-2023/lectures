# Festivos vs Holiday vs vispera

````python
mad_lat = 40.4166852
mad_lon = -3.705739617


# 1. Loop the dataframe
    # if the day: holiday
        # icon diferent
    # marker
    # marker.add_to(map)
    
df.sample()

# How to loop a df?
for index, row in df.iterrows():
    
    # 1. location of the accident
    lat_accident = row["lat"]
    lon_accident = row["lon"]
    
    #dict_coordinates = {"location": [row["lat"], row["lon"]]}
        # Marker(location = [])
    
    if row["festividad"] == "Vispera":
        icon = Icon (
            icon = "glass",
            prefix = "fa",
            color = "red",
            icon_color = "white"
        )
        
    elif row["festividad"] == "Laborable":
        icon = Icon (
                icon = "briefcase",
                prefix = "fa",
                color = "black",
                icon_color = "white"
        )
            
    elif row["festividad"] == "Festivo":
        icon = Icon (
                icon = "television",
                prefix = "fa",
                color = "green",
                icon_color = "white"
        )
    
    new_marker = Marker(location = [lat_accident, lon_accident], icon=icon)
    
    new_marker.add_to(mad_map)
mad_map
````

````python
mad_lat = 40.4166852
mad_lon = -3.705739617
mad_map = Map(location = [mad_lat, mad_lon], zoom_start = 15)

for index, row in df.sample(frac=0.1).iterrows():
    
    all_together = {"location": [row["lat"], row["lon"]]}
            
    # 1. Check for the conditions of Festividad: to get the diferent Icon
    if row["festividad"] == "Laborable":
        icon = Icon (
                icon = "laptop",
                prefix = "fa",
                color = "red",
                icon_color = "white")
        
    elif row["festividad"] == "Vispera":
        icon = Icon (
                icon = "moon-o",
                prefix = "fa",
                color = "white",
                icon_color = "black")
        
    elif row["festividad"] == "Festivo":
        icon = Icon (
                icon = "cutlery",
                prefix = "fa",
                color = "green",
                icon_color = "white")
    
    # 2. With the icon: I pass that to the Marker

    new_marker = Marker(**all_together, icon = icon)
    
    # 3. Add a Marker per row
    
    new_marker.add_to(mad_map)

mad_map
````

##  Morning bikers will be safer  🚴

Let's check how many categories there are in the time column, to make a heat map. First we make a map with the morning schedule, and then we add the afternoon schedule with the Layer Control.
Here we go 🔥

### We create the basic map to which we then add the layers

````python
df["horario"]
````

````python
mad_map = Map(location=[mad_lat, mad_lon], zoom_start=15)
````

````python
mad_map
````

````python
morning = df[df["horario"] == "Mañana"]
evening = df[df["horario"] == "Tarde"]
night = df[df["horario"] == "Noche"]
````

### Subset the dataframe per layer

### We create the FeatureGroup

````python
morning = df[df["horario"] == "Mañana"]
morning_group = folium.FeatureGroup(name=f"Morning: {morning.shape[0]}")
HeatMap(data = morning[["lat", "lon"]]).add_to(morning_group) #only the lat,lon from the mornings
morning_group.add_to(mad_map)
````

````python
mad_map
````

### Create the HeatMap layer and add it to the FeatureGroup

````python
#morning = df[df["horario"] == "Mañana"]
evening_group = folium.FeatureGroup(name=f"Evening: {evening.shape[0]}")
HeatMap(data = evening[["lat", "lon"]]).add_to(evening_group) #only the lat,lon from the mornings
evening_group.add_to(mad_map)

#morning = df[df["horario"] == "Mañana"]
night_group = folium.FeatureGroup(name=f"Night: {night.shape[0]}")
HeatMap(data = night[["lat", "lon"]]).add_to(night_group) #only the lat,lon from the mornings
night_group.add_to(mad_map)
````

````python
mad_map
````
